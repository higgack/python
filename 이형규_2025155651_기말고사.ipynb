```python
# [출제 목적]
# - 두 시점(2025년 1월, 2월) 서울 지하철 이용 데이터를 비교하여
#   데이터 정제, 집계, 비교, 시각화, 변화 분석 등 실무형 데이터 분석 역량을 평가
# - 데이터 전처리, 중복/결측/이상치 처리, 상위 집계, 변화량 비교, 인사이트 도출, 추가 분석 등
# - AI 도구(Perplexity, ChatGPT 등) 활용 시 해당 부분을 주석으로 명시
#
# [AI 도구 활용 내역]
# - 문제 출제와 코드 구조 설계에 Perplexity AI의 아이디어를 참고함
# - pandas, matplotlib, seaborn 활용 예시 및 분석 아이디어 일부 ChatGPT/Perplexity의 예시를 참고하여 재구성함
#
# [과제 요구]
# 1. 2025년 1월과 비교하여 2025년 2월에 새롭게 등장한 역(신규 역) 목록 및 개수 출력
# 2. 2025년 1월에는 있었으나 2025년 2월에는 사라진 역(운영 중단/데이터 미수집 역) 목록 및 개수 출력
# 3. 2025년 1월과 2월 각각 이용객(승차+하차 합산)이 가장 많은 상위 20개 역과 이용객 수를 비교 출력
# 4. 2025년 1월과 2월 각각 상위 100개 역 중 두 달 모두 포함된 역의 리스트와 그 개수 출력
# 5. 2025년 1월 상위 50개 역의 당시 이용객 수와 2월 이용객 수를 비교, 증가/감소 역 파악 및 그래프로 표현
# 6. 위의 통계 도출에 필요한 전처리(중복, 결측, 이상치 등) 과정을 코드와 함께 설명
# 7. 위의 통계 외에도 두 달 변화 분석에 유의미하다고 생각되는 추가 분석을 자유롭게 수행(예: 노선별 변화, 평일/주말 변화 등)하고, 결과를 그래프로 표현 및 해석
# ===============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (윈도우: Malgun Gothic, Mac: AppleGothic)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 1. 파일을 7개 컬럼으로 읽고, 마지막(dummy) 컬럼은 무시
colnames = ["사용일자","노선명","역명","승차총승객수","하차총승객수","등록일자","dummy"]

def load_and_clean(filepath):
    df = pd.read_csv(filepath, encoding='utf-8', names=colnames, header=0, dtype=str)
    df = df.drop(columns=['dummy'])
    df = df.drop_duplicates()
    valid_date_mask = df['사용일자'].str.match(r'^\d{8}$', na=False)
    df = df[valid_date_mask]
    df['사용일자'] = pd.to_datetime(df['사용일자'], format='%Y%m%d', errors='coerce')
    df = df.dropna(subset=['사용일자'])
    df['승차총승객수'] = pd.to_numeric(df['승차총승객수'], errors='coerce').fillna(0).astype(int)
    df['하차총승객수'] = pd.to_numeric(df['하차총승객수'], errors='coerce').fillna(0).astype(int)
    df['총이용객수'] = df['승차총승객수'] + df['하차총승객수']
    return df

df_1 = load_and_clean('CARD_SUBWAY_MONTH_202501.csv')
df_2 = load_and_clean('CARD_SUBWAY_MONTH_202502.csv')

print("2025년 1월 데이터 shape:", df_1.shape)
print("2025년 2월 데이터 shape:", df_2.shape)
print(df_1.head())

# 2. 2월 신규 역 및 1월 사라진 역
stn_1 = set(df_1['역명'].unique())
stn_2 = set(df_2['역명'].unique())
new_stations = sorted(list(stn_2 - stn_1))
disappeared_stations = sorted(list(stn_1 - stn_2))

print("\n[1] 2025년 2월 신규 역 목록 및 개수:")
print(new_stations)
print("신규 역 개수:", len(new_stations))

print("\n[2] 2025년 1월에는 있었으나 2월에는 사라진 역 목록 및 개수:")
print(disappeared_stations)
print("사라진 역 개수:", len(disappeared_stations))

# 3. 각 월별 상위 20개 역 및 이용객수 비교
top20_1 = df_1.groupby('역명')['총이용객수'].sum().sort_values(ascending=False).head(20)
top20_2 = df_2.groupby('역명')['총이용객수'].sum().sort_values(ascending=False).head(20)
print("\n[3] 2025년 1월 상위 20개 역:")
print(top20_1)
print("\n[3] 2025년 2월 상위 20개 역:")
print(top20_2)

# 시각화: 두 달 상위 20개 역 비교
plt.figure(figsize=(16,8))
plt.barh(top20_1.index[::-1], top20_1.values[::-1], alpha=0.7, label='2025-01', color='skyblue')
plt.barh(top20_2.index[::-1], top20_2.values[::-1], alpha=0.4, label='2025-02', color='orange')
plt.title('2025년 1월/2월 상위 20개 역 총 이용객수 비교')
plt.xlabel('이용객수')
plt.legend()
plt.tight_layout()
plt.show()

# 4. 상위 100개 역 중 두 달 모두 포함된 역의 리스트와 개수
top100_1 = set(df_1.groupby('역명')['총이용객수'].sum().sort_values(ascending=False).head(100).index)
top100_2 = set(df_2.groupby('역명')['총이용객수'].sum().sort_values(ascending=False).head(100).index)
common_top100 = sorted(list(top100_1 & top100_2))
print("\n[4] 두 달 모두 상위 100위에 포함된 역 목록:")
print(common_top100)
print("공통 역 개수:", len(common_top100))

# 5. 1월 상위 50개 역의 1월/2월 이용객수 비교, 변화량 시각화
top50_1 = df_1.groupby('역명')['총이용객수'].sum().sort_values(ascending=False).head(50)
top50_names = top50_1.index
top50_2 = df_2.groupby('역명')['총이용객수'].sum().reindex(top50_names).fillna(0)
diff = top50_2 - top50_1
increase = diff[diff > 0]
decrease = diff[diff < 0]

print("\n[5] 1월 상위 50개 역의 1월/2월 이용객수 비교(증가/감소):")
print("증가 역:", increase.index.tolist())
print("감소 역:", decrease.index.tolist())

plt.figure(figsize=(20,8))
plt.bar(top50_names, diff, color=['red' if x < 0 else 'blue' for x in diff])
plt.xticks(rotation=90)
plt.title('2025년 1월 상위 50개 역의 2월 대비 이용객수 변화')
plt.ylabel('2월-1월 이용객수 변화')
plt.tight_layout()
plt.show()

# 6. 전처리 과정 설명
print("""
[6] 데이터 전처리 과정 요약:
- 헤더를 7개로 맞추고 마지막 컬럼(dummy) 무시
- 중복 데이터 제거: drop_duplicates()
- 날짜 컬럼이 8자리 숫자인 행만 남김(이상치 제거): str.match(r'^\\d{8}$')
- 날짜 컬럼 타입 변환: pd.to_datetime
- 승하차 컬럼을 정수형으로 변환: pd.to_numeric(errors='coerce').fillna(0).astype(int)
- 승하차 합산 이용객 컬럼 추가
""")

# 7. 추가 분석1: 노선별 이용객 변화 및 증감률
line_1 = df_1.groupby('노선명')['총이용객수'].sum()
line_2 = df_2.groupby('노선명')['총이용객수'].sum()
lines = sorted(set(line_1.index) | set(line_2.index))
line_1 = line_1.reindex(lines).fillna(0)
line_2 = line_2.reindex(lines).fillna(0)
line_diff = line_2 - line_1
line_rate = (line_diff / line_1.replace(0, np.nan)) * 100

line_compare = pd.DataFrame({
    '2025-01': line_1,
    '2025-02': line_2,
    '증감': line_diff,
    '증감률(%)': line_rate
}).sort_values('증감률(%)', ascending=False)

print("\n[7-1] 노선별 이용객수 및 증감률(%)")
print(line_compare)

plt.figure(figsize=(14,7))
sns.barplot(x=line_compare.index, y=line_compare['증감률(%)'], palette='coolwarm')
plt.title('노선별 2월 대비 1월 이용객수 증감률(%)')
plt.ylabel('증감률(%)')
plt.xlabel('노선명')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 7. 추가 분석2: 평일/주말 이용 변화 (요일 정보 활용)
def add_weekday(df):
    df['요일'] = df['사용일자'].dt.weekday
    df['요일명'] = df['사용일자'].dt.day_name(locale='ko_KR')
    df['주말여부'] = df['요일'].apply(lambda x: '주말' if x >= 5 else '평일')
    return df

df_1 = add_weekday(df_1)
df_2 = add_weekday(df_2)

weekday_1 = df_1[df_1['주말여부'] == '평일'].groupby('역명')['총이용객수'].sum()
weekend_1 = df_1[df_1['주말여부'] == '주말'].groupby('역명')['총이용객수'].sum()
weekday_2 = df_2[df_2['주말여부'] == '평일'].groupby('역명')['총이용객수'].sum()
weekend_2 = df_2[df_2['주말여부'] == '주말'].groupby('역명')['총이용객수'].sum()

# 평일/주말 비율이 큰 역 상위 10개(두 달 평균)
ratio_1 = (weekend_1 / weekday_1.replace(0, np.nan)).sort_values(ascending=False)
ratio_2 = (weekend_2 / weekday_2.replace(0, np.nan)).sort_values(ascending=False)
print("\n[7-2] 2025년 1월 평일대비 주말 이용 비율 상위 10개 역:")
print(ratio_1.head(10))
print("\n[7-2] 2025년 2월 평일대비 주말 이용 비율 상위 10개 역:")
print(ratio_2.head(10))

plt.figure(figsize=(12,6))
plt.bar(ratio_1.head(10).index, ratio_1.head(10).values, alpha=0.7, label='2025-01')
plt.bar(ratio_2.head(10).index, ratio_2.head(10).values, alpha=0.7, label='2025-02')
plt.title('평일대비 주말 이용 비율 상위 10개 역 (두 달 비교)')
plt.ylabel('주말/평일 이용객 비율')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# 7. 추가 분석3: 일자별 전체 이용객 변화 시계열 그래프
daily_1 = df_1.groupby('사용일자')['총이용객수'].sum()
daily_2 = df_2.groupby('사용일자')['총이용객수'].sum()

plt.figure(figsize=(16,5))
plt.plot(daily_1.index, daily_1.values, marker='o', label='2025-01')
plt.plot(daily_2.index, daily_2.values, marker='x', label='2025-02')
plt.title('일자별 전체 지하철 이용객수 (2025년 1월/2월)')
plt.xlabel('날짜')
plt.ylabel('이용객수')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- END ---
```
